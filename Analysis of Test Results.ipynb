# Analysis of Test Results
In this notebook we provide the code and figures of the class-wise accuracy analysis of our prediction results. We will just analyze the best performing approaches, which we indicated at our thesis. 

## Helpdesk

### Tax

```python
# Load prediction Results
tax_base_hd = pd.read_csv('tax\\code\\results\\raw_helpdesk_baseline.txt', sep=';')
tax_orig_hd = pd.read_csv('tax\\code\\results\\raw_helpdesk_orig.txt', sep=';')
tax_dda_hd = pd.read_csv('tax\\code\\results\\raw_helpdesk_cvT_0.95_0.95_dc3_augm.txt', sep=';')

# Plot class-wise accuracy for Tax
plot_class_accuracy(tax_orig_hd, tax_dda_hd, tax_base_hd, 'Helpdesk by Tax')
```

### Mauro

```python
# Load data
nnpm_base_hd = pd.read_csv('nnpm\\results\\raw_helpdesk_baseline.txt', sep=';')
nnpm_orig_hd = pd.read_csv('nnpm\\results\\raw_helpdesk_orig.txt', sep=';')
nnpm_dda_hd = pd.read_csv('nnpm\\results\\raw_helpdesk_cvT_0.95_0.95_dc3_augm.txt', sep=';')

# Map activity categorization to the one of tax
nnpm_orig_hd = map_categorization(df_base=tax_orig_hd, df_adjust=nnpm_orig_hd)
nnpm_dda_hd = map_categorization(df_base=tax_orig_hd, df_adjust=nnpm_dda_hd)
nnpm_base_hd = map_categorization(df_base=tax_orig_hd, df_adjust=nnpm_base_hd)

# Plot class-wise accuracy for mauro
plot_class_accuracy(nnpm_orig_hd, nnpm_dda_hd, nnpm_base_hd, 'Helpdesk by Mauro')
```
